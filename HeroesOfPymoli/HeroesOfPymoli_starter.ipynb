## Age Demographics

* Establish bins for ages


* Categorize the existing players using the age bins. Hint: use pd.cut()


* Calculate the numbers and percentages by age group


* Create a summary data frame to hold the results


* Optional: round the percentage column to two decimal points


* Display Age Demographics Table


```python
# Establish bins for ages
# Categorize the exisiting players using the age bins. Hint: use pd.cut()
# Using 9.9 instead of 10 so that it doesn't inclue 10, 14.9 instead of 15, etc. 

df_age_bins = [0, 9.9, 14.9, 19.9, 24.9, 29.9, 34.9, 39.9, 99999]
df_age_groups = ["<10", "10-14", "15-19", "20-24", "25-29", "30-34", "35-39", "40+"]

df_purchase_data["Age Demographics"]= pd.cut(df_purchase_data["Age"], df_age_bins, labels = df_age_groups)
df_purchase_data
```

```python
# Calculate the numbers and percentages by age group. Similar to gender. 

df_age_groupby = df_purchase_data.groupby("Age Demographics")
df_age_nunique = df_age_groupby["SN"].nunique()
df_age_nunique
```

```python
# Percentage Age count of all players (Male, Female, Other/Non-Disclosed). Similar to gender

df_percentage_age = (df_age_nunique/player_count)*100
df_percentage_age
```

```python
# Create a summary data frame to hold the results
# Optional: give the displayed data cleaner formatting
# Display the summary data frame. Similar to gender

df_age_demographics = pd.DataFrame({"Total Count": df_age_nunique,
                                      "Percentage of Players": df_percentage_age})

df_age_demographics["Percentage of Players"] = df_age_demographics["Percentage of Players"].map("{:,.2f}%".format)

df_age_demographics
```

## Top Spenders

* Run basic calculations to obtain the results in the table below


* Create a summary data frame to hold the results


* Sort the total purchase value column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
# Run basic calculations to obtain SN, purchase count, avg. purchase price, and total purchase value of the top 5 spenders. 

df_SN_groupby = df_purchase_data.groupby("SN")
df_SN_nunique = df_SN_groupby["SN"].nunique()
df_SN_nunique
```

```python
# Run basic calculations to obtain SN, purchase count, avg. purchase price, and total purchase value of the top 5 spenders. 

df_SN_purchase_count = df_SN_groupby["Age"].count()
print(df_SN_purchase_count)

df_SN_avg_purchase_price = df_SN_groupby["Price"].mean()
print(df_SN_avg_purchase_price)

df_SN_total_purchase_value = df_SN_groupby["Price"].sum()
print(df_SN_total_purchase_value)
```

```python
# Create a summary data frame to hold the results
# Sort the total purchase value column in descending order
# Optional: give the displayed data cleaner formatting
# Display the summary data frame

df_top_spenders_analysis = pd.DataFrame({"Purchase Count": df_SN_purchase_count,
                                             "Average Purchase Price": df_SN_avg_purchase_price,
                                             "Total Purchase Value": df_SN_total_purchase_value})

df_top_spenders_analysis = df_top_spenders_analysis.sort_values("Total Purchase Value", ascending=False)


df_top_spenders_analysis["Average Purchase Price"] = df_top_spenders_analysis["Average Purchase Price"].map("${:,.2f}".format)
df_top_spenders_analysis["Total Purchase Value"] = df_top_spenders_analysis["Total Purchase Value"].map("${:,.2f}".format)

df_top_spenders_analysis.head(5)
```

## Most Popular Items

* Retrieve the Item ID, Item Name, and Item Price columns


* Group by Item ID and Item Name. Perform calculations to obtain purchase count, item price, and total purchase value


* Create a summary data frame to hold the results


* Sort the purchase count column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
# Retrieve the Item ID, Item Name, and Item Price columns
# Group by Item ID and Item Name. Perform calculations to obtain purchase count, item price, and total purchase value

df_most_popularitems_groupby = df_purchase_data.groupby(["Item ID", "Item Name"])
df_most_popularitems_groupby.count()

df_popularitems_purchase_count = df_most_popularitems_groupby["Item ID"].count()
print(df_popularitems_purchase_count)

df_popularitems_purchase_price = df_most_popularitems_groupby["Price"].sum()/df_popularitems_purchase_count
print(df_popularitems_purchase_price)

df_popularitems_total_purchase_value = df_most_popularitems_groupby["Price"].sum()
print(df_popularitems_total_purchase_value)
```

```python
# Create a summary data frame to hold the results
# Sort the purchase count column in descending order
# Optional: give the displayed data cleaner formatting
# Display a preview of the summary data frame

df_popularitems_analysis = pd.DataFrame({"Purchase Count": df_popularitems_purchase_count,
                                             "Item Price": df_popularitems_purchase_price,
                                             "Total Purchase Value": df_popularitems_total_purchase_value})

df_popularitems_analysis = df_popularitems_analysis.sort_values("Purchase Count", ascending=False)


df_popularitems_analysis["Item Price"] = df_popularitems_analysis["Item Price"].map("${:,.2f}".format)
df_popularitems_analysis["Total Purchase Value"] = df_popularitems_analysis["Total Purchase Value"].map("${:,.2f}".format)

df_popularitems_analysis.head()
```
